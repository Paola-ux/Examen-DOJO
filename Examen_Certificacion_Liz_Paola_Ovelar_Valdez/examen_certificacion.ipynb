```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importar CSV Salaries

```python
df_salaries = pd.read_csv("C:/Users/sopor/Desktop/Curso Python/archive/ds_salaries.csv")
```

## Limpieza de datos con Python

#### Deteccion y eliminacion de duplicados

```python
#deteccion de datos duplicados
duplicados = df_salaries.duplicated().sum()
print(f"Registros duplicados:{duplicados}")
#eliminacion de datos duplicados
df = df_salaries.drop_duplicates
```

#### Verificacion y ajustes de tipos de datos

```python
print (df_salaries.dtypes)
df_salaries["salary"] = df_salaries["salary"].astype(float)
df_salaries["job_title"] = df_salaries["job_title"].astype("object")
df_salaries ["work_year"] = pd.to_datetime(df_salaries["work_year"])
```

#### Consistencia en valores

```python
df_salaries["job_title"] = df_salaries["job_title"].str.lower().str.strip()
```

#### Identificar y rellenar Valores Faltantes

```python
# Identificar valores faltantes
df_salaries.info()
# Rellenar valores faltantes
df_salaries.fillna(df_salaries["salary"].median(),inplace=True)
df_salaries.fillna({"experience_level":"desconocido"},inplace=True)
```

#### Detectar y corregir datos anomalos

```python
anomalos= df_salaries["salary"]> 1000000
print(f"Datos anomalos:\n{anomalos}")
df_salaries["salary"] = df_salaries["salary"].astype(float)
df_salaries.loc[df_salaries["salary"]> 1000000]=df_salaries["salary"].median()
```

## Exploracion de Datos con Python

### Visualizacion exploratorias univariadas

```python
# Visualizacion 1
fig=plt.figure(dpi=120)
df_salaries.hist("salary")
plt.title("Distribucion de Salario", fontsize=16, fontweight="bold")
plt.xlabel("salary" "\n" "Salario de cada Empleado" )
plt.ylabel("frecuencia" )

# Visualizacion 2
fig=plt.figure(dpi=100)
df_salaries.boxplot("salary")
plt.title("Distribucion de Salario", fontsize=16, fontweight="bold")
plt.xlabel("job_title" "\n" "Puesto que ocupa")
plt.ylabel("experience_level" "\n" "Nivel de experiencia" )
plt.show()
```
